```python
def sizeConverter(value):
    try:
        num,s = value.split(' ')
        return int(num)
    except:
        return np.nan
```

```python
data_cleaning.loc[:,'sizes'] = data_cleaning['size'].apply(sizeConverter)
```

```python
data_cleaning
```

```python
data_cleaning = data_cleaning.drop('size',axis=1)
```

```python
data_cleaning.describe(include= 'all')
```

```python
def sqftConverter(value):
    try:
        if '-' in value:
            num1, num2 = value.split('-')
            return (float(num1)+float(num2))/2
        else:
            return float(value)
    except:
        return np.nan
        
```

```python
data_cleaning['sqft'] = data_cleaning['total_sqft'].apply(sqftConverter)
```

```python
data_cleaning.describe(include='all')
```

```python
data_transform = data_cleaning
```

```python
data_transform = data_transform.dropna(axis=0)
```

```python
features = data_transform[['area_type','sizes','log_sqft','bath','balcony']]
```

```python
target= data_transform['log_price']
features
```

```python
features = features.dropna(axis=0)
features.isnull().sum()
```

```python
target.isnull()
```

```python
features_transform = pd.get_dummies(features,drop_first = True)
```

```python
features_transform.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(features_transform)
```

```python
features_transform = scaler.transform(features_transform)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# List of features matching the number of columns in features_transform
vif = pd.DataFrame()
vif['features'] = ['sizes', 'log_sqft', 'bath', 'balcony', 
                   'area_type_Carpet  Area', 'area_type_Plot  Area', 'area_type_Super built-up  Area']

# Calculate the VIF for all 7 features
vif['collinearity'] = [variance_inflation_factor(features_transform, i) 
                       for i in range(features_transform.shape[1])]

print(vif)
```

```python
type(features_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, ytrain,ytest = train_test_split(features_transform,target,test_size=0.2,random_state=42)
```

```python
x_train, x_test,ytest
```

```python
from sklearn.linear_model import LinearRegression

reg= LinearRegression()
```

```python
reg.fit(x_train,ytrain)
```

```python
reg.score(x_train,ytrain)
```

```python
reg_coffients = pd.DataFrame()
reg_coffients['features']=['sizes', 'log_sqft', 'bath', 'balcony', 'area_type_Carpet  Area',
                                                   'area_type_Plot  Area', 'area_type_Super built-up  Area']
reg_coffients['coffients'] = reg.coef_
```

```python
%%writefile app.py
from flask import Flask, request, jsonify
import pickle
import numpy as np

app = Flask(__name__)

# Load the trained model
model = pickle.load(open("model.pkl", "rb"))

@app.route("/")
def home():
    return "Welcome to the Linear Regression API!"

@app.route("/predict", methods=["POST"])
def predict():
    try:
        data = request.json  # Get the data from frontend

        # Validate and process input features
        required_features = ["sizes", "log_sqft", "bath", "balcony", "area_type_Carpet", "area_type_Plot", "area_type_Super"]
        
        for feature in required_features:
            if feature not in data:
                return jsonify({"error": f"Missing feature: {feature}"}), 400

        input_values = np.array([[data["sizes"], data["log_sqft"], data["bath"], data["balcony"],
                                  data["area_type_Carpet"], data["area_type_Plot"], data["area_type_Super"]]])

        # Make prediction
        prediction = model.predict(input_values).tolist()
        return jsonify({"prediction": prediction[0]})

    except Exception as e:
        return jsonify({"error": str(e)}), 500

if __name__ == "__main__":
    app.run(debug=True)
```
